Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по правилам:


если источник traffic_source равен Yandex или Google, то в source_type ставится organic;

для источников paid и email из России ставим ad;

для источников paid и email не из России ставим other;

все остальные варианты берём из traffic_source без изменений.

```python
import pandas as pd
log = pd.read_csv('visit_log.csv', sep = ';')
log.head()
```

```python
def source_type(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return 'organic'
    elif (row['traffic_source'] == 'paid' or row['traffic_source'] == 'email') and row['region'] == 'Russia':
        return 'ad'
    elif (row['traffic_source'] == 'paid' or row['traffic_source'] == 'email') and row['region'] != 'Russia':
        return 'other'
    else:
        return row['traffic_source']
```

```python
log['source_type'] = log.apply(source_type, axis=1)

log.head(10)
```

Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# Количество оценок для каждого пользователя
ratings_count = ratings.groupby('userId').size()
```

```python
# Фильтрация пользователей, которые выставили более 100 оценок
active_users = ratings_count[ratings_count > 100].index.to_list()
```

```python
# Вычисление времени жизни для каждого пользователя
users_lifetime = []
for user in active_users:
    user_ratings = ratings[ratings.userId == user]
    lifetime = user_ratings.timestamp.max() - user_ratings.timestamp.min()
    users_lifetime.append(lifetime)
```

```python
# Среднее значение времени жизни пользователей
mean_lifetime = sum(users_lifetime) / len(users_lifetime)

print("Среднее время жизни пользователей, которые выставили более 100 оценок: ", mean_lifetime)
```

Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе «Материалы для лекции “Продвинутый pandas”» ---- Ноутбуки к лекции «Продвинутый pandas»).
Нужно сформировать две таблицы:

таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;

аналогичную таблицу по типам выручки с указанием адреса клиента.
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
import pandas as pd
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined = rzd.merge(auto, on='client_id', how='outer')
joined
```

```python
# таблицa с тремя типами выручки для каждого client_id без указания адреса клиента;
joined1 = joined.merge(air, on='client_id', how='outer')
joined1
```

```python
# таблица по типам выручки с указанием адреса клиента
joined2 = client_base.merge(joined1, on='client_id', how='outer')
joined2
```
