Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
# загрузка данных с оценками
ratings = pd.read_csv('ratings.csv')
```

```python
ratings.head()
```

```python
# загрузка данных с названиями фильмов
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
#  присоединение таблиц
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
# фильтрация данных по оценкам 5.0
top_rated = joined[joined['rating'] == 5.0]
top_rated.head()
```

```python
# группировка данных по названию фильма и подсчет количества строк
count_ratings = top_rated.groupby(['movieId', 'title' ])['rating'].count()
```

```python
# определение фильма с наибольшим количеством оценок 5.0
top_rated_movie = count_ratings.idxmax()
print(f'Фильм с наибольшим количеством оценок 5.0: {top_rated_movie}')
```

```python
# СПОСОБ №2 группировка по фильмам и подсчет количества оценок с рейтингом 5.0
joined[joined['rating']==5.0]['movieId'].value_counts().head(1)
```

Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
# Ищем правильное название Латвии
data[ data['country'].str.contains('lat',  case=False) ]['country'].unique()
```

```python
# Ищем правильное название Литвы
data[ data['country'].str.contains('lit',  case=False) ]['country'].unique()
```

```python
# Ищем правильное название Эстонии
data[ data['country'].str.contains('est',  case=False) ]['country'].unique()
```

```python
# фильтруем, оставляя нужные страны
filtered_countries = data[ (data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia') ]
filtered_countries.head()
```

```python
# фильтруем по условию задачи
df_filtered = filtered_countries[(filtered_countries['category']==4) | (filtered_countries['category']==12) | (filtered_countries['category']==21) & (filtered_countries['year']<2010) & (filtered_countries['year']>2005)]
df_filtered.head()
```

```python
print('Количество потребления для стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года равно {}КВт'.format(df_filtered['quantity'].sum()))
```
